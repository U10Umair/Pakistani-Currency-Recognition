### Import Libraries

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Set Dataset Path

```python
base_dir = "/content/drive/MyDrive/Currency Notes detection dataset/Pakistani Currency Dataset"
```

### Resize ONE Image

```python
sample_folder = os.path.join(base_dir, os.listdir(base_dir)[0])
img_name = os.listdir(sample_folder)[0]

img_path = os.path.join(sample_folder, img_name)
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_resized = cv2.resize(img, (224, 224))

plt.imshow(img_resized)
plt.axis("off")
```

### Normalize Image

```python
img_normalized = img_resized / 255.0
```
